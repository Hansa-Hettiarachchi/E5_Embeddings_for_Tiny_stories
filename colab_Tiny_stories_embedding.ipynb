## Creating the data frame from the dataset

Import Libraries

```python
import pandas as pd
import os
```

Use hugging face datsets-server to load Tiny stories dataset and create the dataset

```python
urls = ["https://huggingface.co/datasets/roneneldan/TinyStories/resolve/refs%2Fconvert%2Fparquet/roneneldan--TinyStories/parquet-train-00000-of-00004.parquet",
        "https://huggingface.co/datasets/roneneldan/TinyStories/resolve/refs%2Fconvert%2Fparquet/roneneldan--TinyStories/parquet-train-00001-of-00004.parquet",
        "https://huggingface.co/datasets/roneneldan/TinyStories/resolve/refs%2Fconvert%2Fparquet/roneneldan--TinyStories/parquet-train-00002-of-00004.parquet",
        "https://huggingface.co/datasets/roneneldan/TinyStories/resolve/refs%2Fconvert%2Fparquet/roneneldan--TinyStories/parquet-train-00003-of-00004.parquet"]
df = (
    pd.concat([pd.read_parquet(url) for url in urls])
)
```

Save the dataset in to csv file

```python
csv_file_path = 'Data_Frame.csv'

df.to_csv(csv_file_path,index=False)
```

Extract "text" column from the dataset and put it in a list

```python
text_data = df['text']
text_list = text_data.tolist()
```

```python
print(len(text_list))
```

## E 5 model

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

Load the E5 model from hugging face

```python
model_name = "intfloat/e5-large-v2"

model_kwargs = {'device': 'cuda:0'}


embeddings = HuggingFaceEmbeddings(

    model_name=model_name,

    model_kwargs=model_kwargs

)
```

creating embeddings from tiny stories dataset

```python
embs = embeddings.embed_documents(text_list)
```

```python
df['Embeddings'] = embs
```
